# Multifigure

The Multifigure class allows you to add multiple plots to the same canvas. First, import the library normally and create data to be plotted.

````python
import graphinglib as gl
import numpy as np
````

````python
# Creating data
x_data = np.linspace(-10, 10, 500)
noise = np.random.normal(0, 10, len(x_data))
y_data = 0.05 * x_data ** 3 + x_data ** 2 + x_data + noise

# Creating two Plottable objects
scatter = gl.Scatter(x_data, y_data, label="Scatter plot")
curve = gl.Curve(x_data, y_data, label="Curve plot")
hist = gl.Histogram(y_data, 20, label="Histogram")
````

Next, we will create a Multifigure object to hold the different plots we want to add. The two integers specified are respectively the number of rows and the number of columns for the grid of the multifigure.

````python
multifigure = gl.Multifigure(1, 2, size=(5, 10))
````

To add a plot in our multifigure, we use the ```add_subfigure``` method.

````python
subfigure1 = multifigure.add_subfigure((0, 0, 1, 1))
subfigure1.add_element(scatter)

subfigure2 = multifigure.add_subfigure((0, 1, 1, 1))
subfigure2.add_element(curve)

multifigure.display()
````

The tuple argument contains the following elements: ```(row, column, rows spanned, columns spanned)```. The first two elements are simply the place in the grid where the upper-left corner of the plot will be displayed. The last two elements are the number of rows and columns spanned by the plot. This allows you to fully control the layout of the plots in the canvas.

````python
multifigure2 = gl.Multifigure(2, 4)
subfigure3 = multifigure2.add_subfigure((0, 0, 1, 2))
subfigure3.add_element(scatter)
subfigure4 = multifigure2.add_subfigure((0, 2, 1, 2))
subfigure4.add_element(curve)
subfigure5 = multifigure2.add_subfigure((1, 0, 1, 4))
subfigure5.add_element(hist)
multifigure2.display()
````
