# 2D Plotting

First, import GraphingLib along with Numpy for generating data:

```python
import graphinglib as gl
import numpy as np
```

Next, create Heatmap object to display two-dimensional data:

```python
# Create some data
x, y = np.mgrid[:50, :50]
data = np.cos(x * 0.2) + np.sin(y * 0.3)

# Create the Heatmap object with no parameters
map1 = gl.Heatmap(data)

# Create the Figure object and add the Heatmap to it
fig1 = gl.Figure()
fig1.add_element(map1)

# Display the figure
fig1.display()
```

Some customizations are available as is demonstrated below:

```python
# Create some data
x, y = np.mgrid[:50, :50]
data = np.cos(x * 0.2) + np.sin(y * 0.3)

# Create the Heatmap object with some customizations
map2 = gl.Heatmap(
    data,
    x_axis_range=(0, 1),
    y_axis_range=(0, 1),
    origin_position="lower",
    interpolation="bicubic",
)

# Create the Figure object and add the Heatmap to it
fig2 = gl.Figure()
fig2.add_element(map2)

# Display the figure
fig2.display()
```

It is also possible to provide a path to an image to display using the Heatmap object:

```python
# Create a Heatmap object from an image file
image = gl.Heatmap("images/GraphingLib-Logo-Pixels.png")

# Create the Figure object and add the Heatmap to it
fig3 = gl.Figure()
fig3.add_element(image)

# Display the figure
fig3.display()
```

The color bar shown in the first and second figures is automatically omitted when an image file is used as data. However, every other parameter can be adjusted:

```python
# Create a Heatmap object from an image file with some customizations
image = gl.Heatmap(
    "images/GraphingLib-Logo-Pixels.png",
    x_axis_range=(0, 1),
    y_axis_range=(0, 1),
    interpolation="bicubic",
)

# Create the Figure object and add the Heatmap to it
fig3 = gl.Figure()
fig3.add_element(image)

# Display the figure
fig3.display()
```
